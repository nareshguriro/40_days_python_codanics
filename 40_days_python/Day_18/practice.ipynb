```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the Iris dataset
iris_df = sns.load_dataset('iris') # Replace 'iris.csv' with the actual path to your dataset

# Extract the input and output variables
X = iris_df[['sepal_length', 'sepal_width', 'petal_length']]
y = iris_df['petal_width']

# Create and fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict petal_width based on the input features
y_pred = model.predict(X)

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the actual and predicted values as a scatter plot
ax.scatter(y, y_pred)

# Plot a line between the actual and predicted values
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')

# Set the labels and title
ax.set_xlabel('Actual Petal Width')
ax.set_ylabel('Predicted Petal Width')
ax.set_title('Actual vs Predicted Petal Width')

# Display the plot
plt.show()
```
