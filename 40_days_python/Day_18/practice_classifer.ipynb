### In food likeness dataset no making a classifer model with all those input parameters weight,height, andgender

```python
#Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
df =pd.read_csv('mldata.csv')
df.head()
```

replacing dummy varible male and female with 1 and 0

```python
df['gender']=df['gender'].replace('Male',1)
df['gender']=df['gender'].replace('Female',0)
df.tail(10)
```

```python
#selection of input and output variables
X=df[['age','weight','gender','height']]
y= df['likeness']
```

```python
##Now apply ML algorithm
from sklearn.tree import DecisionTreeClassifier
#create and fit the model
model = DecisionTreeClassifier().fit(X,y)

#prediction 
model.predict([[29,67.0,1,164]])
```

```python
#split data into train and test(80/20)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2) #80% training data and 20% testing data
 
# #create model 
model= DecisionTreeClassifier()
model.fit(X_train,y_train)
predicted_values = model.predict(X_test)
predicted_values

#checking score
score=accuracy_score(y_test,predicted_values)
score
```
